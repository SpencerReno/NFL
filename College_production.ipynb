```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
players = pd.read_csv('./data_files/players.csv')
```

```python
players['collegeName'].value_counts().index[:10]
```

```python
colleges = players['collegeName'].value_counts().index[:10]
values = players['collegeName'].value_counts().values[:10]
```

```python
fig, ax  =plt.subplots(figsize=(20,8))
fig.set_facecolor('#5b626e')
ax.set_facecolor('#5b626e')

sns.barplot(colleges, values, color='skyblue')
plt.title('Best NFL Player Producing Colleges', fontsize=15)
plt.xlabel('colleges', fontsize=14)
plt.ylabel('Players in the 2021 Season', fontsize=14)
plt.show()
```
