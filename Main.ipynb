```python
game_id = 2021090900
play_id = 583

hard_code = all_season[all_season['gameId'] ==game_id]
hard_code = hard_code[hard_code['playId'] ==play_id]
new_game = plays[plays['gameId'] == game_id]
new_game = new_game[new_game['playId'] == play_id].reset_index()

fig, ax  =plt.subplots(figsize=(20,8))
fig.set_facecolor('#5b626e')
ax.set_facecolor('#3B482C')


frame= 0 
plt.scatter(x='x',y='y',data=hard_code[hard_code['frameId'] == 1], 
                        color=hard_code[hard_code['frameId'] == frame]['color'].to_list(), s=100)


def make_ani(test):
    global frame
    frame= frame + 1
    if frame < hard_code['frameId'].unique().shape[0]:
        plt.cla()
        plt.scatter(x='x',y='y',data=hard_code[hard_code['frameId'] == frame], 
                                color=hard_code[hard_code['frameId'] == frame]['color'].to_list(), s=100)


        line_of_scrimage = np.round(hard_code[(hard_code['frameId']==1) & (hard_code['team']=='football')]['x'].reset_index()['x'][0])
        plt.axvline(x=line_of_scrimage, color='blue')
        plt.axvline(x=line_of_scrimage + new_game['yardsToGo'][0], color='Yellow')


        plt.title(f"{new_game['down'][0]} down and {new_game['yardsToGo'][0]} yards to go {new_game['playDescription'][0]}", fontsize=15)
        plt.ylim(0,54)
        plt.yticks([])
        plt.xticks(np.arange(0, 130, 10))
        ax.set_xticklabels(['', 'G', 10 ,20,30,40,50,40,30,20,10, 'G', ''])
        ax.text(3, 13, new_game['defensiveTeam'][0],  rotation=90, fontsize=50)
        ax.text(113, 13, new_game['possessionTeam'][0],  rotation=90, fontsize=50)





        plt.grid(axis='x')


 

ani = animation.FuncAnimation(fig, make_ani, 
                frames=hard_code['frameId'].unique().shape[0],
                interval=50, repeat=False)

video = ani.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()

```
